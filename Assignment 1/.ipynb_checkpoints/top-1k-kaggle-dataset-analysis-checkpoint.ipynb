# 1. Explore Data
In this section, I'll try to extract useful information from the raw data.

## 1.1) Extracts from `last_updated`
### - Estimate data age (assume last update is the data creation)

```python
df['data_age'] = df.last_updated.str.extract(r'([0-9]+.*\s)')
df['data_age']  # This data_age column represents in String dtype. Next, I'll convert to datetime.
```

### - Estimate data creation date

```python
## Parse data_age to constitute creation date in datetime dtype.
units = df.data_age.str.extract(r'(\W.*)')
n = df.data_age.str.extract(r'(\d+)')

## From many units(days, months, years), I'll create Timedelta in 'days' unit to represent data_age.
data_timedelta =  list(zip(n.values.ravel(), units.values.ravel()))

for i, (n_, unit_) in enumerate(data_timedelta):
    
    try:
        n_ = int(data_timedelta[i][0])
    except ValueError as e:
        data_timedelta[i] = (np.nan, 'day')
        continue
    
    if 'month' in unit_:
        n_ *= 30  # assume 1 month = 30 days
        unit_ = 'day'
    if 'year' in unit_: 
        n_ *= 365  # assume 1 year = 365 days
        unit_ = 'day'
    data_timedelta[i] = (n_,unit_.strip())
```

```python
## Create Timedelta, and estimate the creation date
list_data_timedelta = list(map(lambda x: pd.Timedelta(value=x[0], unit=x[1]) ,data_timedelta))
list_data_timedelta = pd.Series(list_data_timedelta)
df['data_age_Timedelta'] = list_data_timedelta

approx_data_creation_date = (pd.to_datetime("today") - list_data_timedelta).astype('datetime64[D]')
df['data_created_date'] = approx_data_creation_date
```

```python
df.data_age_Timedelta = df.data_age_Timedelta.astype('timedelta64[D]')
df.data_created_date = df.data_created_date.astype('datetime64[D]')

df[['data_age_Timedelta', 'data_created_date']]
```

## 1.2) Extract from `files`
### - Data formats distribution

```python
## Extract data formats
files_types = df.files.str.extract(r'(\(.*\))').values.ravel().tolist()
for i, e in enumerate(files_types):
    try:
        types = e[1:-1]
        types = types.split(', ')
        files_types[i] = types
    except TypeError as e:
        files_types[i] = np.nan
        continue
```

```python
## Summarize data formats
files_types_summary = pd.Series(files_types).explode().reset_index(drop = True)
files_types_summary.value_counts().plot(kind='bar', ylabel='number', title='Data formats distribution')
plt.xticks(rotation=0);
```

### - Derive number of files of each dataset

```python
df['number_files'] = df.files.str.extract(r'(\d+)')
```

## 1.3) Extract from `size`
### - Derive total file size of each datasest (in kB)

```python
data_size_kb = df['size'].str.split().values.tolist()
assert df['size'].isnull().sum() == 0

bigQ_data = dict()
for i, size in enumerate(data_size_kb):
    
    try:
        n_, unit_ = size
    except ValueError as e:
        bigQ_data[i]=size[0]
        continue
        
    n_ = int(n_)
    if 'm' in unit_.lower():
        n_ *= 1024  # 1 MB = 1024 KB
    elif 'g' in unit_.lower():
        n_ *= (1024*1024)  # 1 GB = 1024 MB
    data_size_kb[i] = n_
    
df['data_size_kb'] = data_size_kb
```

```python
df.number_files = pd.to_numeric(df.number_files, downcast='integer', errors='coerce')
df.data_size_kb = pd.to_numeric(df.data_size_kb, downcast='integer', errors='coerce')
df['avg_size_per_file'] = df.data_size_kb / df.number_files
```

## 2.3) Analysis

```python
col_used = ['usability', 'upvotes', 'data_age_Timedelta', 'number_files', 'data_size_kb']
## I don't include badge since badge is basically a dependent variable as well.

df_analyze = df[col_used]
df_analyze = df_analyze.fillna(value=df_analyze.median())  ## Impute NaN with Median due to huge skew

X = df_analyze.drop(['upvotes'], axis=1)
Y = df_analyze['upvotes']
```

```python
fig, axes = plt.subplots(1,4, figsize=(15,4), sharey=True)
fig.suptitle('data distribution')

for i, col in enumerate(X):
    X[col].plot(kind='hist', ax=axes[i])
    axes[i].set_xlabel(col)
```

```python
def Normalize(column, isLog=False, isSqrt=False):
    if isLog:
        column = np.log(column + 1)
    elif isSqrt:
        column = np.sqrt(column)
    upper = column.max()
    lower = column.min()
    y = (column - lower)/(upper-lower)
    return y

fig, axes = plt.subplots(1,4, figsize=(15,4), sharey=True)
fig.suptitle('Log-normalized')

for i, col in enumerate(X.columns):
    Normalize(X[col], isLog=True).plot(kind='hist', ax=axes[i])
    axes[i].set_xlabel(col)
```

Even when being applied Log-transformation, the data still highly skewed. So, we shouldn't use linear model to find the feature importance. However, let's give it a try.

```python
import statsmodels.api as sm

X_norm = X.apply(Normalize, axis=0, isSqrt=True)

X_norm = sm.add_constant(X_norm) # adding a constant
Y_norm = Normalize(Y)
model = sm.OLS(Y_norm, X_norm).fit()
print_model = model.summary()
print(print_model)
```

R-squared is only 0.016 meaning that, with the linear model, only 1.6% of predictors can actually cover the variation of the dependent variable.

```python
from sklearn import linear_model
clf = linear_model.Lasso(alpha=0.1)
clf.fit(X_norm, Y_norm)

print(clf.coef_)
print(clf.intercept_)
```

Since Lasso gives zero coef for all variables, we are ensured that, according to the dataset we have, we cannot use linear model to find feature importance.

```python
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor()
model.fit(X, Y)

rank = list(zip(X.columns, np.round(model.feature_importances_, 3)))
sorted(rank, key = lambda x: x[-1], reverse=True)
```

By using feature_importance from Random forest, we can see that the size of data is the most relevent feature for determining the number of upvotes.

Next, we'll see the popular topic among the top 1000 Kaggle datasets.

```python
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import nltk
from nltk.corpus import stopwords
import re
import string

stop_words = '|'.join(stopwords.words('english'))
text = ' '.join(df.title).upper()
text = ''.join([e for e in text if e not in string.punctuation])
text = re.sub(rf"\s(dataset|data|prediction|image|classification|new|{stop_words})\w*", "", text, flags=re.IGNORECASE)

wordcloud = WordCloud(width = 800, height = 800, background_color ='white').generate(text)

plt.figure( figsize=(7,7) )
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
words = pd.Series(text.split())
print('Top 10 words in datasets:')
print(words.value_counts()[:10])
```
