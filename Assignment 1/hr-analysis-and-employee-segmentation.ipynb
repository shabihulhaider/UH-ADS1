```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.style.use('ggplot')

orange_black = ['#fdc029', '#df861d', 'FF6347', '#aa3d01',
                '#a30e15', '#800000', '#171820']

plt.rcParams['figure.figsize'] = (10,5) 
plt.rcParams['figure.facecolor'] = '#FFFACD' 
plt.rcParams['axes.facecolor'] = 'FFFFE0' 
plt.rcParams['axes.grid'] = True 
plt.rcParams['grid.color'] = orange_black[3]
plt.rcParams['grid.linestyle'] = '--' 
```

```python
df = pd.read_csv('HRdata.csv').iloc[:,1:]
df.head()
```

```python
df.describe().T
```

```python
ageCut = pd.cut(df.Age, [0,20,35,45,65],labels=['18-20','21-35','36-45','46-65']).value_counts().reset_index()
ageCut.columns = ['Age Range','Count']
ageCut.sort_values('Age Range').reset_index(drop=True).style.bar(color='#d65f5f')
```

```python
plt.figure()

sns.scatterplot(data = df, x='Age', y='Salary',hue='Status', edgecolor='k')
plt.show()

sns.violinplot(data=df, x='Status', y='Age')
plt.show()
```

```python
plt.figure(figsize=(14,8))

sns.scatterplot(data = df, x='Age', y='Salary',hue='Region', edgecolor='k')
plt.show()

sns.violinplot(data=df, x='Region', y='Age')
plt.show()
```

```python
plt.figure(figsize=(14,8))

sns.scatterplot(data = df, x='Age', y='Salary',hue='Gender', edgecolor='k')
plt.show()

sns.violinplot(data=df, x='Gender', y='Age')
plt.show()
```

```python
plt.figure(figsize=(14,8))

sns.scatterplot(data = df, x='Age', y='Salary',hue='Marital', edgecolor='k')
plt.show()

sns.violinplot(data=df, x='Marital', y='Age')
plt.show()
```

```python
plt.figure(figsize=(15,4))
plt.subplot(1,3,1)
sns.kdeplot(df[df['Status']=='Blue-Collar']['Age'], label='Blue-Collar')
sns.kdeplot(df[df['Status']=='White-Collar']['Age'], label='White-Collar')
plt.title('Age')
plt.legend()
plt.subplot(1,3,2)
sns.kdeplot(df[df['Status']=='Blue-Collar']['Senior'], label='Blue-Collar')
sns.kdeplot(df[df['Status']=='White-Collar']['Senior'], label='White-Collar')
plt.title('Seniority')
plt.legend()
plt.subplot(1,3,3)
sns.kdeplot(df[df['Status']=='Blue-Collar']['Salary'], label='Blue-Collar')
sns.kdeplot(df[df['Status']=='White-Collar']['Salary'], label='White-Collar')
plt.title('Salary')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(15,4))
sns.countplot(data = df, x='Department', hue='Status')
plt.xticks(rotation=90)
plt.ylabel('')
plt.show()
```

```python
plt.figure(figsize=(15,4))
sns.countplot(data = df.sort_values('Education Level'), x='Education Level', hue='Status')
plt.xticks(rotation=90)
plt.ylabel('')
plt.show()
```
